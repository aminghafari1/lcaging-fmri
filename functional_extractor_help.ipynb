```python
import nibabel as nib
import os    
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import clear_output
```

```python
filename='fMRI_avg_brain.nii.gz'
current_dir = os.getcwd()
image_address = os.path.join(current_dir, filename)
img = nib.load(filename)
data=img.get_fdata()
```

```python
print(np.shape(data))
```

```python
## SHowing the image
for i in range(np.shape(data)[0]):
    plt.imshow(data[i,:,:], cmap='gray')
    plt.show()
    ## pause for 1 second
    plt.pause(0.25)
    ## clear whatever output that we have
    clear_output(wait=True)



```

```python
slices=data[60,:,:]
plt.imshow(slices, cmap='gray')
plt.show()
```

```python
slices=data[60,:,:]
plt.imshow(slices, cmap='gray')
plt.show()
editted_slices=slices
threshold=130
```

```python
print("Welcome to this cell.")
for i in range(np.shape(editted_slices)[0]):
    for j in range(np.shape(editted_slices)[1]):
        if slices[i][j]>threshold:
            break
        else:
            editted_slices[i][j]=0
```

```python
plt.imshow(editted_slices, cmap='gray')
plt.show()
```
